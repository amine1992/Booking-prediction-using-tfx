## Preprocess input

```python
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)
import sys
sys.path.append('../scripts')
import preprocess
import os
import tensorflow_data_validation as tfdv
```

```python
TRAIN_DATA_DIR = "../data/train"
EVAL_DATA_DIR = "../data/eval"
TFT_TRAIN_FILE_PREFIX = 'train_transformed'
TFT_EVAL_FILE_PREFIX = 'eval_transformed'
```

```python
!rm -R -f ../data/train/bookings_output
# Transform training data
preprocess.transform_data(input_handle=os.path.join(TRAIN_DATA_DIR, 'train.csv'),
                          outfile_prefix=TFT_TRAIN_FILE_PREFIX, 
                          working_dir=os.path.join(TRAIN_DATA_DIR,'bookings_output'),
                          schema_file='../data/tfdv_output/schema.pbtxt',
                          pipeline_args=['--runner=DirectRunner'])
print('Done')
```

```python
!rm -R -f ../data/eval/bookings_output
# Transform eval data
preprocess.transform_data(input_handle=os.path.join(EVAL_DATA_DIR, 'eval.csv'),
                          outfile_prefix=TFT_EVAL_FILE_PREFIX, 
                          working_dir=os.path.join(EVAL_DATA_DIR,'bookings_output'),
                          schema_file='../data/tfdv_output/schema.pbtxt',
                          pipeline_args=['--runner=DirectRunner'])
print('Done')
```

```python
# Compute stats over transformed training data.
TRANSFORMED_TRAIN_DATA = os.path.join(os.path.join(TRAIN_DATA_DIR,'bookings_output'), TFT_TRAIN_FILE_PREFIX + "*") 
transformed_train_stats = tfdv.generate_statistics_from_tfrecord(data_location=TRANSFORMED_TRAIN_DATA)
```

```python
tfdv.visualize_statistics(transformed_train_stats)
```
